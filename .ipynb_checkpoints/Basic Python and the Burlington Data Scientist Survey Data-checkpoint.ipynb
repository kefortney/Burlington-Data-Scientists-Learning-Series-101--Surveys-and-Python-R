## Columns

Because the orginal survey had questions as headers the column names are quite long. Changing them to something easier to work with will make everything easier. The columns function can both return what columns are there and help you rename them.

```python
# Returns column names, in jupyter the print command is not necessary, but may be a good habit to have
print(survey.columns)
```

```python
# renames the columns to the new list of values
survey.columns = ['id', 'start', 'end', 'describe', 'degree', 'job', 'technical_demo', 'training_methods', 'big_projects', 
  'solving_challenges', 'social_events', 'R', 'Python', 'stats_software', 'Tableau', 'spreadsheets', 'C++', 
  'SQL', 'Hadoop', 'Java', 'Hive', 'Pig', 'Ruby', 'Perl', 'percent_cleaning', 'percent_analyzing data', 
  'percent_visualizing', 'percent_databases', 'meetups_attended', 'connected_with_colleagues', 'job_or_project', 
  'new_skills', 'recruited', 'participated_other_meetups', 'training', 'Mon 6-7:30', 'Mon 7:30-9', 'Tue 6-7:', 
  'Tue 7:30-', 'Wed 6-7:30', 'Wed 7:3', 'Thu 6-7:3', 'Thu 7:30-9']
      
```

## Dates and times

More often than not you have to deal with dates and times. Both the Panda's datetime function and the datetime library are useful for navigating the data type and allows for intuitive plotting in Matplotlib and Seaborn.

```python
# Sets the start and end time of the survey to a datetime data type
survey['start'] = pd.to_datetime(survey['start'])
survey['end'] = pd.to_datetime(survey['end'])

# Splitting out the time and the date is easy with a lambda function and the datetime library, it will add columns to the far right
survey['timeonly'],survey['dateonly']= survey['start'].apply(lambda x:x.time()), survey['start'].apply(lambda x:x.date())
```

## Data Manipulation

Often in surveys the data is returned in a form that is categorical (string or text) versus numeric so conversion is needed. The approach may very on how the survey was done, in this case the values range from "Definetly Very Interested" to "Definetly Not Interested" so either that can be represented as 1 to 5 or -2 to 2 depending on your approach.

Likewise use of different languages can be represented as a range depending on the return. Once these values can be numerically represented a correlation (not the same as causation) can be determined with with numeric values from -1 (negative correlation) to a +1 range. Above 0.5 is usually significant depending on the question.

```python
# Returns all the unique values for a given column
pd.unique(survey['big_projects']) 
```

```python
# Creates a dictionary paring expected string and numerical value
di = {'Definitely\xa0VERY interested':5, 'Somewhat interested':4, 'Neutral':3,
      'NOT interested':2, 'Definitely\xa0NOT interested':1}

# Replaces all NaN strings and empmpy NAN
survey = survey.replace('NaN', np.NAN)
survey = survey.replace('', np.NAN)

# Replaces values based on the dictionary for a given column
survey['training_methods'].replace(di, inplace=True)
survey['technical_demo'].replace(di, inplace=True)
survey['big_projects'].replace(di, inplace=True)
survey['solving_challenges'].replace(di, inplace=True)
survey['social_events'].replace(di, inplace=True)
```

```python
pd.unique(survey['R'])
```

```python
# Creates a dictionary paring expected string and numerical value
di = {'Never use':1,'Rarely\xa0use':2,'Regularly use':3,
       'Use all the time':4}

# Replaces values based on the dictionary for a given column
survey['R'].replace(di, inplace=True)
survey['Python'].replace(di, inplace=True)
survey['stats_software'].replace(di, inplace=True)
survey['spreadsheets'].replace(di, inplace=True)
survey['C++'].replace(di, inplace=True)
survey['SQL'].replace(di, inplace=True)
survey['Hadoop'].replace(di, inplace=True)
survey['Java'].replace(di, inplace=True)
survey['Hive'].replace(di, inplace=True)
survey['Pig'].replace(di, inplace=True)
survey['Ruby'].replace(di, inplace=True)
survey['Perl'].replace(di, inplace=True)
```

```python
# This is a way to cut the hour data off the dataset be selection just the specified columns, remember that python is zero-index
surveymod = survey.iloc[:,0:35]
surveymod.tail()
```

```python
corr = surveymod.corr()
corr
```

```python
sns.heatmap(corr)
plt.show()
```

## Seaborn Visualizations

Seaborn can help you explore your data, in this case we will run through a couple of chart types given the modification of the data we have done.

```python
sns.lmplot(y='R',x='training_methods', hue='degree', data=survey)
plt.title('R Language and Learning Training Methods')
plt.xlabel('Training Methods')
plt.ylabel('Use of R Language')
plt.xticks(rotation=45)
plt.show
```

```python
sns.lmplot(y='Python',x='training_methods', hue='degree', data=survey)
plt.title('Python Language and Learning Training Methods')
plt.xlabel('Training Methods')
plt.ylabel('Use of PYthon')
plt.xticks(rotation=45)
plt.show
```
