## Dates and times

More often than not you have to deal with dates and times. Both the Panda's datetime function and the datetime library are useful for navigating the data type and allows for intuitive plotting in Matplotlib and Seaborn.

```python
# Sets the start and end time of the survey to a datetime data type
survey['start'] = pd.to_datetime(survey['start'])
survey['end'] = pd.to_datetime(survey['end'])
```

## Data Transformation

Often in surveys the data is returned in a form that is categorical (string or text) versus numeric so conversion is needed. The approach may very on how the survey was done, in this case the values range from "Definetly Very Interested" to "Definetly Not Interested" so here I represent it as a range of 1 to 5.

```python
# Returns all the unique values for a given column
pd.unique(survey['big_projects']) 
```

```python
# Creates a dictionary paring expected string and numerical value
di = {'Definitely\xa0VERY interested':5, 'Somewhat interested':4, 'Neutral':3,
      'NOT interested':2, 'Definitely\xa0NOT interested':1}

# Replaces all NaN strings and empmpy NAN
survey = survey.replace('NaN', np.NAN)
survey = survey.replace('', np.NAN)

# Replaces values based on the dictionary for a given column
survey['training_methods'].replace(di, inplace=True)
survey['technical_demo'].replace(di, inplace=True)
survey['big_projects'].replace(di, inplace=True)
survey['solving_challenges'].replace(di, inplace=True)
survey['social_events'].replace(di, inplace=True)
```

```python
# Returns all the unique values for a given column
pd.unique(survey['R'])
```

```python
# Creates a dictionary paring expected string and numerical value
di = {'Never use':1,'Rarely\xa0use':2,'Regularly use':3,
       'Use all the time':4}

# Replaces values based on the dictionary for a given column
survey['R'].replace(di, inplace=True)
survey['Python'].replace(di, inplace=True)
survey['stats_software'].replace(di, inplace=True)
survey['spreadsheets'].replace(di, inplace=True)
survey['C++'].replace(di, inplace=True)
survey['SQL'].replace(di, inplace=True)
survey['Tableau'].replace(di, inplace=True)
survey['Hadoop'].replace(di, inplace=True)
survey['Java'].replace(di, inplace=True)
survey['Hive'].replace(di, inplace=True)
survey['Pig'].replace(di, inplace=True)
survey['Ruby'].replace(di, inplace=True)
survey['Perl'].replace(di, inplace=True)
```

## Visualization

Matplotlib (https://matplotlib.org/) is the common visualization tool, Seaborn (https://seaborn.pydata.org/) is built on top of it and one can utilize matplotlib and seaborn together. Most commonly matplotlib is imported as plt and seaborn as sns.  

```python
# A basic histogram plot of what hour the survey was completed
plt.hist(survey['start'].dt.hour, bins=40)
plt.show()
```

## Seaborn Visualizations

Seaborn can help you explore your data, in this case we will run through a couple of chart types given the modification of the data we have done.

```python
sns.distplot(survey['solving_challenges'])
sns.distplot(survey['social_events'])
plt.title('Distribution of Rating for Solving Challenges to Social Events')
plt.xlim(0,5)
plt.xlabel('Distribution')
plt.show
```

```python
sns.barplot(y='job', x='Tableau', data=survey)
plt.title('Occupation and Use of Tableau')
plt.ylabel('')
plt.show()
```

```python
sns.violinplot(x='solving_challenges',y='job', data=survey)
plt.title('Distribution of importance of Solving Challenges and Job Type')
plt.xlim(0,5)
plt.show()
```

```python
sns.regplot(y='percent_cleaning',x='percent_analyzing',  data=survey)
plt.title('Linear Regression of Percent of Time Cleaning vs Analyzing')
plt.xlabel('Percent of Time Analyzing')
plt.ylabel('Percent of Time Cleaning')
plt.show()
```

```python
sns.lmplot(y='Hadoop',x='technical_demo', hue='degree', data=survey)
plt.title('Use of Hadoop and Interested in Technical Demonstrations')
plt.xlabel('Training Methods')
plt.ylabel('Use of R Language')
plt.xticks(rotation=45)
plt.show
```
